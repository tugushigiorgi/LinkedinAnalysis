```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Setting Indexing   Company_id Column both of Dataframes
JobPostings=pd.read_csv('job_postings.csv',index_col="company_id",parse_dates=["listed_time"])
Companyinfo=pd.read_csv("company_details/companies.csv",index_col="company_id")
```

```python
#Lets see what Columnts are there 
JobPostings.columns
Companyinfo.columns 
print(JobPostings.columns)
print(Companyinfo.columns )

```

```python
#See Top 10 Company who posted most in that date
#Merging Files Because Jobposting Containts only Job id and not Company name , Company name is in CompanyInfo csv
merged_data = JobPostings.merge(Companyinfo, on="company_id")
merged_data.rename(columns={"name":"Company"}, inplace=True)


top_10_companies = merged_data['Company'].value_counts().head(10)
top_10_companies
```

```python
#see Top 10 Country posts most

merged_data["country"].value_counts().head(10)
merged_data
```

```python
#see All jobs posted in Usa
usafiltration=merged_data["country"]=="US"
 
merged_data.loc[usafiltration,["Company","title"]]
 
```

```python
#Filtration :see USA jobs which is Full time only


usafulltimefiltration=(merged_data["country"]=="US") & (merged_data["work_type"]=="FULL_TIME")
 
merged_data.loc[usafulltimefiltration,["Company","title"]]
```

```python
#See Highest paying Job Title 
highest_salary_job = merged_data[merged_data['max_salary'] == merged_data['max_salary'].max()]
highest_salary_job
```

```python

#Number of Applications  Fulltime vs Contract:
fulltime_avg_applications = merged_data[merged_data['formatted_work_type'] == 'Full-time']['applies'].sum()
contract_avg_applications = merged_data[merged_data['formatted_work_type'] == 'Contract']['applies'].sum()
print(f"Fulltime Post : {fulltime_avg_applications}")
print(f"Contract Post : {contract_avg_applications}")
 
```

```python
#average salary by experience

average_salary_by_experience = merged_data.groupby('formatted_experience_level')['min_salary'].mean().reset_index()

#and  Sorting  by average salary in ascending order
average_salary_by_experience = average_salary_by_experience.sort_values(by='min_salary')

average_salary_by_experience
```

```python
#sorting by comapny_size and Company Name 

SortedwithsizeandName=merged_data.sort_values(by=["company_size","Company"],ascending=[False,True])

SortedwithsizeandName
```

```python
#mean , min , max standart dev, median;
meanofSalary=merged_data["max_salary"].mean()
MinOfSalary=merged_data["max_salary"].min()
MaxofSalary=merged_data["max_salary"].max()
salarystd=merged_data["max_salary"].std()
salarymedian=merged_data["max_salary"].median()
print(f"Mean of Salay : {meanofSalary}")
print(f"Min of Salary : {MinOfSalary}")
print(f"Max of Salary :{MaxofSalary}")
print(f"Std of Salary :{salarystd}")
print(f"Median of Salary {salarymedian}")
    
```

```python

#Bar diagram with Average Applications by Work Type (Contract, Full time,internship etc..)
bar_data = merged_data.groupby('formatted_work_type')['applies'].mean()
bar_data.plot(kind='bar')
plt.title('Average Applications by Work Type')
plt.xlabel('Work Type')
plt.ylabel('Average Applications')
plt.show()
```

```python
line_data = merged_data.groupby('formatted_experience_level')['job_id'].count()

#Line diagram with Job Postings by Experiance Level (Associate,Director,Entry level,Interntship etc..)
line_data.plot(kind='line')
plt.title('Job Postings by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Number of Job Postings')
plt.xticks(rotation=45)  
plt.show()
```

```python
 
#Another Line chart 

# Group and count job postings by location (top 10)
line_data =merged_data['location'].value_counts()
top_locations = line_data.head(10)
top_locations.plot(kind='line', marker='o', linestyle='-', color='b', markersize=8)
plt.title('Job Postings by Location (Top 10)')
plt.xlabel('Location')
plt.ylabel('Number of Job Postings')
plt.xticks(rotation=45)   
plt.grid(True)
plt.show()
 
```
